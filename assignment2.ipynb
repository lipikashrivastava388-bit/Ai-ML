# Linear Regression Model on Housing Dataset predicting price

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## Load Dataset

```python
dataset = pd.read_csv('Housing.csv')
dataset.head()
```

```python
dataset.info()
```

## Convert Categorical Variables

```python
yes_no_cols = ['mainroad','guestroom','basement',
               'hotwaterheating','airconditioning','prefarea']

for col in yes_no_cols:
    dataset[col] = dataset[col].map({'yes':1, 'no':0})
```

```python
if 'furnishingstatus' in dataset.columns:
    dataset = pd.get_dummies(dataset, columns=['furnishingstatus'], drop_first=True)
```

```python
X = dataset.drop('price', axis=1)
y = dataset['price']
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Feature Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Build Linear Regression Model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

## Model Evaluation

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print("MAE :",mae)
print("MSE :",mse)
print("RMSE :",rmse)
print("R^2 Score:",r2)
```

```python
coeff_dr = pd.DataFrame(lr.coef_,X.columns, columns=['Coefficient'])
coeff_dr.sort_values(by='Coefficient',ascending=False)
```

# Logistic Regression on Titanic dataset for survival prediction

## Data Wrangling

```python
# libraries already imported
titanic_data = pd.read_csv('Titanic-Dataset.csv')
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.drop("Cabin",axis=1,inplace=True)
titanic_data.head()
```

```python
titanic_data.dropna()
```

```python
Sex = pd.get_dummies(titanic_data['Sex'],drop_first=True)
Sex.head()
```

```python
Embarked = pd.get_dummies(titanic_data['Embarked'],drop_first=True)
Embarked.head()
```

```python
Pcl = pd.get_dummies(titanic_data['Pclass'],drop_first=True)
Pcl.head(5)
```

```python
titanic_data = pd.concat([titanic_data,Pcl,Embarked,Sex],axis=1)
titanic_data.head()
```

## Train Data

```python
X = titanic_data.drop('Survived', axis=1)
y = titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel = LogisticRegression(max_iter=1000, solver="liblinear")
logmodel.fit(X_train.to_numpy(), y_train.to_numpy())
```

```python
predictions = logmodel.predict(X_test.to_numpy())
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
